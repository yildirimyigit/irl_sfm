```python
desired_len = 400 # min_len
x_dim, gamma_dim, y_dim = 2, 2, 2

all_X = np.zeros((num_tests, num_demos, desired_len, x_dim))
all_gamma =  np.zeros((num_tests, num_demos, desired_len, gamma_dim))
all_Y = np.zeros((num_tests, num_demos, desired_len, y_dim))

for k in all_demonstrations:
    print(f'k:{k}')
    X, gamma, Y = np.zeros((num_demos, desired_len, x_dim)), np.zeros((num_demos, desired_len, gamma_dim)), np.zeros((num_demos, desired_len, y_dim))
    for i, demo in enumerate(all_demonstrations[k]):
        len_demo = len(demo)
        indices = np.linspace(0, len_demo-1, desired_len, dtype=int)  # <desired_len>-many indices are selected uniformly
        x, gmm, y = np.zeros((desired_len, x_dim)), np.zeros((desired_len, gamma_dim)), np.zeros((desired_len, y_dim))
        for new_ind, old_ind in enumerate(indices):
            x[new_ind] = demo[old_ind][:x_dim]
            gmm[new_ind] = demo[old_ind][x_dim:x_dim+gamma_dim]
            y[new_ind] = demo[old_ind][-y_dim:]
        X[i] = x
        gamma[i] = gmm
        Y[i] = y

    p_path = f'{processed_data_path}{k}/'
    np.save(p_path+'d_x.npy', X)
    np.save(p_path+'d_gamma.npy', gamma)
    np.save(p_path+'d_y.npy', Y)
    
    if not k == 'novel':
        all_X[int(k)] = X
        all_gamma[int(k)] = gamma
        all_Y[int(k)] = Y
    else:
        all_X[-1] = X
        all_gamma[-1] = gamma
        all_Y[-1] = Y
        np.save(combined_data_path+'novel/d_x.npy', X)
        np.save(combined_data_path+'novel/d_gamma.npy', gamma)
        np.save(combined_data_path+'novel/d_y.npy', Y)

print('DONE')
```

```python
num_val = 3 # num of runs for validation

out_data_path = '/home/yigit/phd/yigit_phd_thesis/cnmp/data/sfm/robot_obs_1/combined/'

# reshaping to meet CNMP requirements
# X = np.reshape(all_X, (num_demos, desired_len, x_dim))
# gamma = np.reshape(all_gamma, (num_demos, desired_len, x_dim))
# Y = np.reshape(all_Y, (num_demos, desired_len, y_dim))

x_train, x_val, y_train, y_val, gamma_train, gamma_val = [], [], [], [], [], []

val_indices = np.sort(np.random.choice(range(num_tests-1), num_val, replace=False))
for i in range(num_tests):
    if not i == num_tests-1:
        for j in range(num_demos):
            if not np.abs(all_X[i][j][0][0]) > 0.5:  # discard paths with starting positions other than x=0
                if i in val_indices:  # validation instance
                    print(f'val: {i}-{j}')
                    x_val.append(all_X[i][j])
                    gamma_val.append(all_gamma[i][j])
                    y_val.append(all_Y[i][j])
                else:  # training instance
                    print(f'train: {i}-{j}')
                    x_train.append(all_X[i][j])
                    gamma_train.append(all_gamma[i][j])
                    y_train.append(all_Y[i][j])
    else:
        for j in range(num_demos):
            if not np.abs(all_X[i][j][0][0]) > 0.5:
                print(f'nov: {i}-{j}')
                np.save(out_data_path + 'novel/' + 'd_x.npy', np.array(all_X[i][j].reshape(1,400,2)))
                np.save(out_data_path + 'novel/' + 'd_y.npy', np.array(all_Y[i][j].reshape(1,400,2)))
                np.save(out_data_path + 'novel/' + 'd_gamma.npy', np.array(all_gamma[i][j].reshape(1,400,2)))

np.save(out_data_path + 'demonstrations/' + 'd_x.npy', np.array(x_train))
np.save(out_data_path + 'demonstrations/' + 'd_y.npy', np.array(y_train))
np.save(out_data_path + 'demonstrations/' + 'd_gamma.npy', np.array(gamma_train))
np.save(out_data_path + 'demonstrations/' + 'v_d_x.npy', np.array(x_val))
np.save(out_data_path + 'demonstrations/' + 'v_d_y.npy', np.array(y_val))
np.save(out_data_path + 'demonstrations/' + 'v_d_gamma.npy', np.array(gamma_val))
```
